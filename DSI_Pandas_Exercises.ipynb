```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3/2, label="y = x**3/2")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

```python
# Update the matplotlib configuration parameters: Using the STIX fonts
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
#Important LaTex note regarding back slashes
#Matplotlib has great support for LaTeX. All we need to do is to use dollar signs encapsulate
#LaTeX in any text (legend, title, label, etc.). For example, "$y=x^3$".
#But here we can run into a slightly subtle problem with LaTeX code and Python text strings 
#In LaTeX, we frequently use the backslash in commands, for example \alpha to produce the symbol α. 
#But the backslash already has a meaning in Python strings (the escape code character). 
#To avoid Python messing up our latex code, we need to use "raw" text strings. 
#Raw text strings are prepended with an 'r', like r"\alpha" or r'\alpha' instead of "\alpha" or '\alpha':

fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$") #Using latex
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title');
```

```python
x = np.linspace(0,10,30)
dy = 0.1
y = np.random.normal(np.sin(x)*np.sin(x), dy)

fig, ax = plt.subplots()
plt.errorbar(x,y,dy,fmt = '.k')
```

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where=(y1<y2), color = 'red')
```

```python
x = np.random.random(50)
y=np.random.random(50)
c = 50*np.random.random(50)
s = 500* np.random.random(50)

fig, ax = plt.subplots()
im = ax.scatter(x,y,c=c,s=s,cmap = plt.cm.RdBu)
#adding a color bar
fig.colorbar(im, ax=ax)
im.set_clim(0,50.0)
```

```python
x = np.linspace(0,10,50)
y = np.linspace(0,20,60)
z = np.cos(y[:,np.newaxis])*np.sin(x)
fig, ax = plt.subplots()

im = ax.contourf(x,y,z,100)
im2 = ax.contour(x,y,z,colors='k')
fig.colorbar(im, ax=ax)
```

```python
I = np.random.random((100,100))
I+= np.cos(np.linspace(0,3*np.pi,100))
fig, ax = plt.subplots()
im = ax.imshow(I, cmap = plt.cm.PuOr)
fig.colorbar(im, ax=ax)
```
